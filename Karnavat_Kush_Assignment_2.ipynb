```python
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
d1 = df[["abbrev",'total','speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses']].sort_values(by="ins_premium").tail(10)
d2 = df[["abbrev",'total','speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses']].sort_values(by='ins_losses').tail(10)
d3 = df[["abbrev","ins_premium"]].sort_values(by="ins_premium").head(10)
d4 = df[["abbrev",'total','speeding','alcohol','not_distracted','no_previous','ins_premium','ins_losses']].sort_values(by='total').head(10)
```

```python
ax = sns.scatterplot(x="ins_losses", y="ins_premium", data=df, s=40)
```

```python
ax = sns.scatterplot(x=d2["no_previous"], y=d2["ins_losses"],hue=d2["abbrev"])
```

```python
ax = sns.scatterplot(x=d4["no_previous"], y=d4["ins_premium"],hue=d4["abbrev"])
```

```python
#Scatter plot shown between ins_premium and ins_losses

ax = sns.scatterplot(x=d4["ins_premium"], y=d4["ins_losses"],hue=d4["abbrev"])
```

```python
#Relplot between total and ins_premium
sns.relplot(x="total",y="ins_premium",data=d4,hue="abbrev")
```

```python
#Plotting a bar plot of the top 10 states with highest car insurance premium

plt.figure(figsize=(12,10))
sns.barplot(x=d1['abbrev'], y=d1['ins_premium'],palette=sns.cubehelix_palette(len(d1['abbrev'])))
plt.xticks(rotation= 90)
plt.xlabel('ABBREV')
plt.ylabel('ins_premium')
plt.title('Costliest car_insurance_premiums top 10 states')
plt.show()
```

```python
#Plotting a bar plot of the top 10 states with highest losses incurred by insurance companies

plt.figure(figsize=(12,10))
sns.barplot(x=d2['abbrev'], y=d2['ins_losses'],palette=sns.cubehelix_palette(len(d2['abbrev']),start=0.5,rot=-0.75))
plt.xticks(rotation= 90)
plt.xlabel('ABBREV')
plt.ylabel('ins_losses')
plt.title('Top 10 states with maximum ins_losses')
plt.show()
```

```python
#Plotting a bar plot of the top 10 states with lowest car insurance premium

plt.figure(figsize=(12,10))
sns.barplot(x=d3['abbrev'], y=d3['ins_premium'],palette=sns.cubehelix_palette(len(d3['abbrev'])))
plt.xticks(rotation= 90)
plt.xlabel('ABBREV')
plt.ylabel('ins_premium')
plt.title('cheapest ins_premium top 10 states')
plt.show()
```

```python
#Plotting a bar plot of the top 10 states with least number of drivers involved in fatal collisions

plt.figure(figsize=(12,10))
sns.barplot(x=d4['abbrev'], y=d4['total'],palette=sns.cubehelix_palette(len(d4['abbrev'])))
plt.xticks(rotation= 90)
plt.xlabel('ABBREV')
plt.ylabel('TOTAL')
plt.title('Top 10 states with least TOTAL')
plt.show()
```
